```python
import matplotlib.pyplot as plt
import pandas as pd

year = [1, 2, 3, 4, 5]
revenue = [78, 93.6, 112.3, 134.8, 161.7]
cost = [36.5, 43.8, 52.6, 63.1, 75.7]
profit = [41.5, 49.8, 59.7, 71.7, 86]

data = pd.DataFrame({'year': year, 'revenue': revenue, 'cost': cost, 'profit': profit})

plt.plot(data['year'], data['revenue'], label='Doanh thu', color='blue')
plt.plot(data['year'], data['cost'], label='Chi phí', color='red')
plt.plot(data['year'], data['profit'], label='Lợi nhuận', color='green')

plt.title('Dự báo doanh thu cho FreshMeal')
plt.xlabel('Năm')
plt.ylabel('Tỷ đồng')
plt.legend(title='Chỉ tiêu')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dữ liệu
price = [40000, 45000, 50000, 55000, 60000]
profit = [26, 33.75, 41.5, 49.25, 57]
data = pd.DataFrame({'price': price, 'profit': profit})

# Tạo biểu đồ
plt.figure(figsize=(8, 6))  # Kích thước của biểu đồ

# Vẽ đường
plt.plot(data['price'], data['profit'], color='blue', marker='o', linestyle='-')

# Thêm điểm trên đường
plt.scatter(data['price'], data['profit'], color='red', label='Điểm dữ liệu')

# Đặt tiêu đề và nhãn trục
plt.title('Phân tích độ nhạy của lợi nhuận theo đơn giá', fontsize=16)
plt.xlabel('Đơn giá (đồng)', fontsize=14)
plt.ylabel('Lợi nhuận (tỷ đồng)', fontsize=14)

# Thêm chú thích
for i, txt in enumerate(data['profit']):
    plt.annotate(f'{txt} tỷ', (data['price'][i], data['profit'][i]), textcoords="offset points", xytext=(0, 5), ha='center')

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.grid(True)
plt.show()
```
