Dự báo doanh thu là quá trình ước tính doanh thu của một công ty trong một khoảng thời gian nhất định, dựa trên các dữ liệu lịch sử, xu hướng hiện tại, và các yếu tố thị trường. Phân tích độ nhạy là quá trình xem xét sự thay đổi của các chỉ tiêu hiệu quả tài chính khi các yếu tố có liên quan đến chỉ tiêu đó thay đổi.
Dự báo doanh thu và phân tích độ nhạy cho FreshMeal, với các thông tin:
•	Đơn giá của một gói ăn sẵn của FreshMeal (P)
•	Số lượng người tiêu dùng tiềm năng trong thị trường dẫn đầu (Q)
•	Tần suất sử dụng sản phẩm (F)
•	Tỉ lệ chuyển đổi khách hàng ©
•	Chi phí biến đổi (VC)
•	Chi phí cố định (FC)
•	Tỉ lệ tăng trưởng doanh thu (G)
Giả sử bạn có các thông tin sau:
•	P = 50.000 đồng
•	Q = 2.000.000 người
•	F = 3 lần/tuần
•	C = 10%
•	VC = 30.000 đồng/gói
•	FC = 5 tỷ đồng/năm
•	G = 20%/năm

```python
import matplotlib.pyplot as plt
import pandas as pd

year = [1, 2, 3, 4, 5]
revenue = [78, 93.6, 112.3, 134.8, 161.7]
cost = [36.5, 43.8, 52.6, 63.1, 75.7]
profit = [41.5, 49.8, 59.7, 71.7, 86]

data = pd.DataFrame({'year': year, 'revenue': revenue, 'cost': cost, 'profit': profit})

plt.plot(data['year'], data['revenue'], label='Doanh thu', color='blue')
plt.plot(data['year'], data['cost'], label='Chi phí', color='red')
plt.plot(data['year'], data['profit'], label='Lợi nhuận', color='green')

plt.title('Dự báo doanh thu cho FreshMeal')
plt.xlabel('Năm')
plt.ylabel('Tỷ đồng')
plt.legend(title='Chỉ tiêu')
plt.show()
```

Giả sử chúng ta muốn xem xét ảnh hưởng của việc thay đổi đơn giá (P) lên lợi nhuận (P) của FreshMeal trong năm đầu tiên. Có thể giả định rằng các yếu tố khác như Q, F, C, VC, FC, G không thay đổi. Chúng ta có thể thay đổi P từ 40.000 đồng đến 60.000 đồng, với bước nhảy là 5.000 đồng, và tính P tương ứng như bên dưới

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dữ liệu
price = [40000, 45000, 50000, 55000, 60000]
profit = [26, 33.75, 41.5, 49.25, 57]
data = pd.DataFrame({'price': price, 'profit': profit})

# Tạo biểu đồ
plt.figure(figsize=(8, 6))  # Kích thước của biểu đồ

# Vẽ đường
plt.plot(data['price'], data['profit'], color='blue', marker='o', linestyle='-')

# Thêm điểm trên đường
plt.scatter(data['price'], data['profit'], color='red', label='Điểm dữ liệu')

# Đặt tiêu đề và nhãn trục
plt.title('Phân tích độ nhạy của lợi nhuận theo đơn giá', fontsize=16)
plt.xlabel('Đơn giá (đồng)', fontsize=14)
plt.ylabel('Lợi nhuận (tỷ đồng)', fontsize=14)

# Thêm chú thích
for i, txt in enumerate(data['profit']):
    plt.annotate(f'{txt} tỷ', (data['price'][i], data['profit'][i]), textcoords="offset points", xytext=(0, 5), ha='center')

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.grid(True)
plt.show()
```
